```python
site_list = ['Detroit', 'Columbus', 'Green Bay','Springfield', 'Omaha']
size_list = ['S1', 'S2', 'S3', 'S4', 'S5']
flavor_list = ['F1', 'F2', 'F3', 'F4', 'F5', 'F6', 'F7', 'F8', 'F9', 'F10', 'F11', 'F12']
PFO_ = {}
for i in site_list:
    PFO_[i] = {}
    for j in size_list:
        PFO_[i][j] = {}
        for k in flavor_list:
            a = PFO[(PFO.Site == i) & (PFO.Size == j) & (PFO.Flavor == k)].Processing_Rate
            PFO_[i][j][k] = a
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig = plt.figure(figsize= (10,8))
sns.violinplot(packaging[(packaging.Site == 'Detroit') &
         (packaging.Size == 'S1') & 
         (packaging.Packaging_Type == 'Box')].Processing_Rate)
plt.show()
```

```python
pack_list = ['Box', 'Bag']
pack = {}
for i in site_list:
    pack[i] = {}
    for j in size_list:
        pack[i][j] = {}
        for k in pack_list:
            pack[i][j][k] = packaging[(packaging.Site == i) & 
                                      (packaging.Size == j) & 
                                      (packaging.Packaging_Type == k)].Processing_Rate.mean()
```
